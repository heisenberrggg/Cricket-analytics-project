```python
import pandas as pd
import json
```

## Match Summary Transformation

```python
with open('Downloads/t20_json_files/t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)
```

```python
df_match= pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.rename({'scorecard': 'match_id'},axis=1,inplace=True)
df_match.head()
```

```python
csv_file = 't20_wc_match_results.csv'
df_match.to_csv(csv_file, index=False)
```

## Batting Summary Transformation

```python
with open('Downloads/t20_json_files/t20_json_files/t20_wc_batting_summary.json') as b:
    data = json.load(b)
```

```python
all_rec=[]
for rec in data:
   all_rec.extend(rec['battingSummary'])

df_batting=pd.DataFrame(all_rec)
```

```python
df_batting.head()
```

```python
df_batting['out/not_out']=df_batting['dismissal'].apply(lambda x: 'out' if len(x)>0 else 'not out')
df_batting.head()
```

```python
df_batting.drop(['dismissal'],axis=1,inplace=True)
df_batting.head(25)
```

```python
import re

def remove_special_characters(text):
    return re.sub(r'[^A-Za-z0-9 ()]+', '', text)

df_batting['batsmanName']=df_batting['batsmanName'].apply(remove_special_characters)
df_batting.head(10)
```

```python
df_match.head(10)
```

```python
dict={}

for index,row in df_match.iterrows():
    key1 = row['team1']+' Vs '+row["team2"]
    key2 = row['team2']+' Vs '+row["team1"]
    dict[key1]=row['match_id']
    dict[key2]=row['match_id']

dict
```

```python
df_batting['match_id']=df_batting['match'].map(dict)
df_batting.head(10)
```

```python
csv_file = 't20_wc_batting_summary.csv'
df_batting.to_csv(csv_file, index=False)
```
